# El costo del financiamiento

En este proyecto realizamos una simulación de el costo que tendrá pagar la universidad teniendo un financiamiento de 30,000 pesos.
Es decir, si el semestre costará 100,000 pesos se toma 30% de financiamiento. Y eso es lo que lo dejamos a futuro para pagarlo terminando la universidad, suponiendo que terminamos en 3 años y medio.

### Objetivo general

Simular lo que se tendría que pagar al final de nuestros 7 semestres restantes, por un financiamiento de 30,000 pesos al semestre.

### Objetivos específicos

- Trasladar los 30,000 pesos al valor de las Udis actual.
- Simular el valor de las udis hasta el 1ero de Mayo del 2022.
- Convetir el valor de las udis simuldas a cantidad de dinero.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Consulta.csv',index_col='Fecha')
data.plot()
```

```python
pct = data.pct_change().dropna()
pct
```

```python
# Graficamos los rendimientos que ha tenido el precio de UDIs en el tiempo desde 1995.
pct.plot(grid=True)
```

Parece ser que en los inicios de las Unidades de Inversion los rendimientos que tenían eran muy altos, para fines de no alterar demasiado la simulación se procederá a eliminar los primeros datos. hasta que se vea una tendencía 'normal'.

```python
# Graficamos el periodo que se pretende a eliminar 
# debido a que tiene rendimientos muy altos en comparación con el resto de la gráfica. 
pct.head(1500).plot(grid=True)
```

```python
pct = pct.tail(7151)
pct
```

```python
# Se grafica el restante
pct.plot(grid=True)
```

```python
pct.plot(figsize=(20,4),grid=True)
```

```python
mean = pct.mean().values[0]
mean
```

```python
std = pct.std().values[0]
std
```

```python
UDI = data.tail(1).values[0][0]
UDI
```

```python
# Comprobamos que los rendimientos tienen un comportamiento normal.
plt.hist(pct.values,bins=80);
```

```python
dias = 182
simulaciones = 100000

Sim = np.random.randn(dias, simulaciones)*std + mean
```

```python
last = data.tail(1).values[0][0]
last
```

```python
his = (np.cumprod(Sim+1, axis=0)*last)
```

```python
plt.plot(his);
```

```python
pred = his[-1]
plt.hist(pred, bins=40);
```

```python
prom = pred.mean()
stdv = pred.std()
prom, stdv
```

```python
pred.sort()
aprox = (pred[int(.975*simulaciones)]-pred[int(.025*(simulaciones))])/2
prom - aprox, prom + aprox
```

```python
# queremos calcular la cantidad aproximada a pagar al finalizar la carrera, aproximadamente 3 años y medio. 
# 14 de enero de 2019 - 1 de mayo de 2022 (1203 días.)
dias = 1203
simulaciones = 100000

Sim = np.random.randn(dias, simulaciones)*std + mean
last = data.tail(1).values[0][0]
his = (np.cumprod(Sim+1, axis=0)*last)
pred = his[-1]
prom = pred.mean()
stdv = pred.std()
pred.sort()
aprox = (pred[int(.975*simulaciones)]-pred[int(.025*(simulaciones))])/2
pred_min , pred_max = prom - aprox, prom + aprox
pred_min, pred_max
```

El precio de UDIs el 1 de mayo de 2022 será aproximadamente de 7.1994 pesos +/- 0.07868, con una probabilidad del 95%

```python
# supongamos que hoy debemos $30,000 pesos en financiamiento. Estos en udis serían: 
udis = 30000/last
# después de 3 años y medio será hora de pagar el financiamiento, por tanto pagar la cifra en udis. 
udis*pred_min,udis*pred_max
```

Al finalizar la carrera se deberá pagar entre 34494.75 y 35257.06 pesos por un financiamiento adquirido 3 años y medio antes con una confianza cercana al 95%.
